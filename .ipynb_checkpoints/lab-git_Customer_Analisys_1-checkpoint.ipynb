```python
clv_new = []
for item in data['customer_lifetime_value']:
    if pd.isna(item):
        clv_new.append(np.nan)
    elif type(item) == str and '%' in item:
        clv_new.append(float(item[:-1]))
    else:
        clv_new.append(item)
```

```python
data['customer_lifetime_value'] = clv_new
```

```python
data['customer_lifetime_value'] = pd.to_numeric(data['customer_lifetime_value'], errors='coerce')
```
